```python
import numpy as np
import matplotlib.pyplot as plt
import sympy

from IPython import display
```

Run xpt to understand gradient descent.

```python
#Dfn the Function.
x = np.linspace(-2*np.pi, 2*np.pi,401)
fx = np.sin(x)* np.exp(-x**2*0.05)

#Get the Derivative.
df = np.cos(x)*np.exp(-x**2*.05)+ np.sin(x)*(-.1*x)*np.exp(-x**2*0.5)

#quick (inspection calls.)
plt.plot(x,fx,x,df)
plt.legend(['f(x)','df'])
plt.show()
```

```python
#dfn the fxn
def fuxn(x):
    return np.sin(x)* np.exp((-x**2)*0.05)

def deriv(x):
    # x= sympy.Symbol('x')
    # df = np.sin(x) * np.exp(-x**2*0.05)
    df = np.sin(x)*np.exp((-x**2)*0.05)*.1*x + np.exp((-x**2)*0.05)*np.cos(x)
    # display.display(df)
    return df

#train.
local_minimum = np.random.choice(x,1)
print(local_minimum)
learning_rate = 0.01
training_epochs = 1002

for i in range(training_epochs):
    gradient = deriv(local_minimum)
    local_minimum = local_minimum - gradient * learning_rate

#plot them out for visualisation.

x = np.linspace(-5,5,50)
plt.plot(x,fuxn(x),x,deriv(x))
plt.legend(['fxn','deriv'])
plt.xlim(x[[0,-1]])
plt.plot(local_minimum,fuxn(local_minimum),'ro')
plt.plot(local_minimum,deriv(local_minimum),'go')
plt.show()
# deriv(-5)
```

```python
#Expt 1. systematically vary the starting locations.

start_locs = np.linspace(-5,5,50)
final_results = np.zeros(len(start_locs))

for idx, local_minimum in enumerate(start_locs):
    for i in range(training_epochs):
        gradient = deriv(local_minimum)
        local_minimum = local_minimum - learning_rate * gradient

    final_results[idx] = local_minimum

#plot the results
plt.plot(start_locs, final_results,'s-')
plt.xlabel('Starting guess')
plt.ylabel('Final guess')
plt.show()
```
