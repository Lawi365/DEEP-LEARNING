```python
#The Peaks function.

def peaks(x,y):
    x,y = np.meshgrid(x,y)

    z = 3*(1-x)**2 * np.exp(-(x**2)) - (y+1)**2 \
        -10 * (x/5 - x**3 - y**5) * np.exp(-x**2-y**2)\
            -1/3*np.exp(-(x+1)**2 - y**2)
    # display(z)
    return z

#create a landscape
x = np.linspace(-3,3,201)
y= np.linspace(-3,3,201)

z = peaks(x,y)

plt.imshow(z)
plt.show()
```

```python
#we then create the derivative using sympy.Abs

def der_peak(x,y):
    return sympy.diff(peak(x,y))
```
