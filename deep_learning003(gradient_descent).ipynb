```python
import torch 
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
# display.set_matplotlib_formats('svg')
```

```python
# Gradient Descent in 1D
# create a function to do this.
#for this example we are hard coding this.
def fxn(x):
    return 3*x**2 - 3*x + 4

#creating the derivative function.
def derivative(x):
    return 6*x -3

#plot the function and part of the derivative.
#define the x values.
x = np.linspace(-2,2,2001)

#plot them
plt.plot(x, fxn(x), x, derivative(x))
plt.xlim(x[[0,-1]])
plt.legend(['fxn','derivative'])
plt.show()
```

```python
#Creating a gradient Descent algol
# create the random starting point.
# np.random.seed(98)
local_minimum = np.random.choice(x,1)
print(local_minimum)
#learing parameters.
learning_rate = 0.01
training_epochs = 1001

#now train
for i in range(training_epochs):
    # //first get the derivative.
    # //then at that derivative get the gradient at that local point local_minimum
    # //then adjust the local_minimum till global is reached
    gradient = derivative(local_minimum)
    # printing to get an idea
    print(f'local_minimum %g, gradient %g'%(local_minimum,gradient))
    local_minimum -= learning_rate * gradient


print(local_minimum)

#Then we plot the results for Further Understanding.
plt.plot(x, fxn(x), x, derivative(x))
plt.plot(local_minimum, derivative(local_minimum),'go')
plt.plot(local_minimum, fxn(local_minimum), 'ro')
plt.show()
```

```python
model_parameters = np.zeros((training_epochs,2))

local_minimum = np.random.choice(x,1)
print(local_minimum)
#learing parameters.
learning_rate = 0.01
training_epochs = 1001

#now train
for i in range(training_epochs):
    # //first get the derivative.
    # //then at that derivative get the gradient at that local point local_minimum
    # //then adjust the local_minimum till global is reached
    gradient = derivative(local_minimum)   
    local_minimum -= learning_rate * gradient
    model_parameters[i,:] = local_minimum,gradient

#plot the gradient over iterations
fig, ax = plt.subplots(1,2,figsize=(12,4))

for i in range(2):
    ax[i].plot(model_parameters[:,i], 'o-')
    ax[i].set_xlabel('Iteration')
    ax[i].set_title(f'Final estimated minimum: {local_minimum[0]:.5f}')

ax[0].set_ylabel('Local minimum')
ax[1].set_ylabel('Derivative')

plt.show()
```
